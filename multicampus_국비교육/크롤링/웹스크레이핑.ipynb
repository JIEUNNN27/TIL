# 네이버 뉴스 크롤링
네이버 뉴스 검색 "무역전쟁"을 3페이지를, 다음과 같은 형태로 저장.


```python
# [{  
#     'url': 'https://..',  
#     'title': '',  
#     'desc': '',  
#     'contents': '', # 상세페이지 html
#     'thumbnail': '' # 뉴스 목록에 있는 이미지 없을시 None
# }]
```

2. 검색결과 url에 요청보내기
3. BeautifulSoup으로 parsing하고 저장하기
4. (1~3)을 3페이지까지 반복시키기

```python
url = 'https://search.naver.com/search.naver'
params = {
    'where': 'news',
    'query': '무역전쟁'
}
resp = requests.get(url, params=params)
resp
```

```python
soup = BeautifulSoup(resp.text)
sample = soup.select('.bx')
print(len(sample))
# 태그까지 같이 명시.
sample2 = soup.select('li.bx')
print(len(sample2))
```

```python
sample3 = soup.select('#sp_nws1')
print(len(sample3))
# 태그까지 같이 명시.
sample4 = soup.select('li#sp_nws1')
print(len(sample4))
```

```python
# parsing할거에요.
# [{  
#     'url': 'https://..',  
#     'title': '',  
#     'desc': '',  
#     'thumbnail': '', # 뉴스 목록에 있는 이미지 없을시 None
#     'contents': '', # 상세페이지 html
# }]
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver'
params = {
    'where': 'news',
    'query': '무역전쟁'
}
result_list = []

for i in range(3):
    params['start'] = i*10 + 1 
    resp = requests.get(url, params=params)
    soup = BeautifulSoup(resp.text)

    news_list = soup.select('ul.list_news > li.bx')
    len(news_list)

    
    for news in news_list:
        img_tag = news.select_one('.dsc_thumb > img.thumb')
        img = None
        if img_tag:
            img = img_tag.get('src')
        
        news_dict = {
            'url': news.select_one('a.news_tit').get('href'),
            'title': news.select_one('a.news_tit').text,
            'desc': news.select_one('.dsc_txt_wrap').text,
            'thumbnail': img 
        }
        result_list.append(news_dict)
```

```python
for item in result_list:
    resp = requests.get(item['url'])
    item['contents'] = resp.text
```

```python
len(result_list), result_list
```

```python
news_result_list = result_list.copy()
```
